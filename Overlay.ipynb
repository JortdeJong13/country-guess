# Overlay 
Overlay the actual country shape over the drawing

```python
# Imports
import matplotlib.pyplot as plt
from countryguess.data import TestDataset, Dataset, poly_to_img
```

```python
test_data = TestDataset(shape=(196, 196))
ref_data = Dataset(shape=(196, 196))
```

```python
idx=20
```

```python
sample = test_data[idx]

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(sample["drawing"], interpolation='nearest', origin='lower')
axs[0].set_title(f"User drawing of {sample['country_name']}") 

img = poly_to_img(ref_data.from_country_name(sample['country_name']), ref_data.shape)
axs[1].imshow(img, interpolation='nearest', origin='lower')
axs[1].set_title("Reference shape")
plt.show()
```

## Naive overlay

```python
sample = test_data[idx]

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(sample["drawing"], interpolation='nearest', origin='lower')
axs[0].set_title(f"User drawing of {sample['country_name']}") 

img = poly_to_img(ref_data.from_country_name(sample['country_name']), ref_data.shape)
axs[1].imshow(img + sample["drawing"] > 0, interpolation='nearest', origin='lower')
axs[1].set_title("Reference shape")
plt.show()
```
